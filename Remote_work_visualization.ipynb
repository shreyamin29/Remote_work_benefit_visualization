```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = [10, 6]


filename = '/content/Benefits of remote work.xlsx'
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_excel(filename)
df.head(50)
```

```python

df_ben = pd.DataFrame({'Categories of Benefits':['Ability to have a flexible schedule', 'Flexibility to work from any location', 'Not having to commute',
                                                 'Ability to spend time with family', 'Ability to work from home', 'Other'], 'Percentage':[32,26,21,11,7,3]})
                  

ax = df_ben.plot.bar(x='Categories of Benefits', y='Percentage', rot=0, color=['green', 'red', 'yellow', 'blue', 'pink', 'purple'])


plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 15)
#plt.yticks(rotation =0)
plt.ylim(0, 38)
#plt.xlim(0, 30)
plt.xlabel('Categories of Benefits', fontsize = 12)
plt.ylabel('Percent', fontsize = 12)
plt.title('Benefits of Remote Work', fontsize = 15);
plt.show()
```
